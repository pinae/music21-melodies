# Notes with different names in the Circle of Fifths

The Circle of Fifths is usually constructed with sharps where needed on the right side and flats on the left side. In order to get these keys by raising by a fifth one has to start with D flat as the seventh note like this:

```python
from music21.note import Note
from music21.stream import Stream

base_note = Note('C4', quarterLength=4)
notes = [base_note]
for i in range(6):
    new_note = base_note.transpose('P5')
    while new_note.octave > 4:
        new_note = new_note.transpose('P-8')
    notes.append(new_note)
    base_note = new_note
base_note = Note('D4-', quarterLength=4)
notes.append(base_note)
for i in range(4):
    new_note = base_note.transpose('P5')
    while new_note.octave > 4:
        new_note = new_note.transpose('P-8')
    notes.append(new_note)
    base_note = new_note
s = Stream(notes)
s.show()
s.show('midi')
```

In the code this looks like something not very elegant. One should be able to calculate the Circle of Fifths by just adding a fifth by each step like this:

```python
base_note = Note('C4', quarterLength=4)
notes = [base_note]
for i in range(10):
    new_note = base_note.transpose('P5')
    while new_note.octave > 4:
        new_note = new_note.transpose('P-8')
    notes.append(new_note)
    base_note = new_note
s2 = Stream(notes)
s2.show()
s2.show('midi')
```

This certainliy sounds like the Circle of Fifths but the notation only uses sharps. With sharps instead of flats the notation looks quite different. Does the differing notation mean that the notes are different? Let's hear:

```python
s3 = Stream()
for i in range(11):
    s3.append(s.notes[i])
    s3.append(s2.notes[i])
s3.show()
s3.show('midi')
```

Did you hear any difference?

Comparing the notes one to one by ear the pairs sound exactly the same. This shows that there are different ways to write the same note.

One could write all notes by only using sharps or only using flats. However you would need 11 shaprs for a song in D minor while a sigle flat would be enough. With G major or E minor it is the other way around: 1 sharp or 11 flats.

While producing notes with exactly the same pitch music21 saves and shows the ambiguous notes with sharps and a one higher note with flat differently. In order to still get the notation usually used with the Circle of Fifths it is better to calculate the notes in two parts with the second part stating with D flat.
